```
from sklearn.datasets import make_blobs
from matplotlib import pyplot as plt

random_state = 42
color_map = {0: 'red', 1: 'blue'}
```

```
# Problem 1: Using sklearn
X, y = make_blobs(n_samples = 250, centers = [(1, 1), (-1, -1)], cluster_std = 1/8, random_state = random_state)
```

```
colors = [color_map[label] for label in y]
plt.scatter(X[:, 0], X[:, 1], c = colors, s = 10, alpha = .5)
plt.show()
```

```
# Problem 2: Using numpy.random.randn
import numpy as np

np.random.seed(random_state)
```

```
X = np.array(list(zip(1/8 * np.random.randn(125) + -1, 1/8 * np.random.randn(125) + -1)) + list(zip(1/8 * np.random.randn(125) + 1, 1/8 * np.random.randn(125) + 1)))
y = ([1] * 125) + ([0] * 125)
colors = [color_map[label] for label in y]
plt.scatter(X[:, 0], X[:, 1], c = colors, s = 10, alpha = .5)
plt.show()
```

```
# Problem 3: Less separated data
X, y = make_blobs(n_samples = 250, centers = [(1, 1), (-1, -1)], cluster_std = 4, random_state = random_state)
plt.scatter(X[:, 0], X[:, 1], c = colors, s = 10, alpha = .5)
plt.show()
```

```
# Problem 4: Distinct but not linearly separable data
from sklearn.datasets import make_circles

X, y = make_circles(n_samples = 500, random_state = random_state)
X = X[y == 0] # Filter out other circle

# Make inner blob
X_blob, _ = make_blobs(n_samples = 250, centers = [(0, 0)], cluster_std = 1/8, random_state = random_state)
X = np.concatenate((X, X_blob), axis = 0)

y = np.concatenate((y[y == 0], np.array([1] * 250)), axis = 0)

colors = [color_map[label] for label in y]

plt.scatter(X[:, 0], X[:, 1], s = 10, alpha = .5, c = colors)
plt.show()
```
